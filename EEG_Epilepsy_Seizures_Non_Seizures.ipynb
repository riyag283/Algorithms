```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('epilepsy.csv')
```

```
X = df.values
X = X[:,1:-1]
y = np.array(df['y'])
samples = len(X)
dict1 = {}
dict1[1] = 'red'
dict1[2] = 'lightblue'
dict1[3] = 'lightblue'
dict1[4] = 'lightblue'
dict1[5] = 'lightblue'
```

```
plt.figure(figsize=(12,8))
for i in range(samples//500):
    plt.plot(X[i,:],color=dict1[y[i]])
plt.show()
```
